1. Создайте массив numpy.
Мы можем создавать массивы:
- из списков 
arr = np.range([1,2,3])
- из кортежей
arr = np.range((1,2,3))
- задаваемые последовательностью
arr = np.arange(10) # выводит 10 цифр от 0 до 9
c границами 
arr = np.arange(10, 20) # выводит 10 цифр от 10 до 19
с границами и шагом
arr = np.arange(2, 10, 3) # выводит цифры от 2 до 10(не включая) с шагом 3
